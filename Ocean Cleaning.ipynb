```python
import pandas as pd 
import numpy as np
data = "Ocean.csv"
df = pd.read_csv(data)

df.head()
```

```python
df.tail()
```

```python
df.columns
print(df.dtypes)
```

```python
df.drop(['CO2 SW QF', 'CO2 Air QF','H2O SW (mmol/mol)', 'dfCO2','dpCO2'],axis = 1)
```

```python
df.to_csv('Ocean data.csv', index=False)
```

```python
df.replace(-999.0, np.nan, inplace=True)
```

```python
columns_to_fill = ['pH (total scale)', 'xCO2 SW (wet) (umol/mol)', 'xCO2 Air (wet) (umol/mol)', 
                   'xCO2 SW (dry) (umol/mol)', 'xCO2 Air (dry) (umol/mol)', 
                   'fCO2 SW (sat) uatm', 'fCO2 Air (sat) uatm', 
                   'pCO2 SW (sat) uatm', 'pCO2 Air (sat) uatm']

for column in columns_to_fill:
    if column in df.columns:  # Ensure the column exists in the DataFrame
        mean_value = df[column].mean(skipna=True)
        df[column] = df[column].fillna(mean_value)
```

```python
df.tail()
```

```python
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:  # Check if column is numeric
        df[column].fillna(df[column].mean())
```

```python
df.tail()
```

```python
print(df.isnull().sum())
```

```python
df.to_csv('Ocean data.csv', index=False)
```

```python
def estimate_pH(fCO2):
    if fCO2 > 0:  # Ensures valid values
        return 8.4 - 0.1 * (fCO2 / 1000)  # Example: Adjust constants as needed
    return np.nan

# Apply the function to your dataset
df['pH (total scale)'] = df['fCO2 SW (sat) uatm'].apply(lambda x: estimate_pH(x) if not pd.isna(x) else np.nan)

# Replace remaining NaN values with the mean
df['pH (total scale)'].fillna(df['pH (total scale)'].mean())

# Save or preview the result
print(df[['fCO2 SW (sat) uatm', 'pH (total scale)']])
```

```python
df.tail()
```

```python
df.to_csv('Ocean data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is loaded in a DataFrame `df`
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')  # Convert to datetime
df['Year'] = df['Date'].dt.year  # Extract the year from the Date column

# Plotting the pH (total scale) over the years
plt.figure(figsize=(10, 6))

# Aggregate data by Year (you can choose other aggregations like mean, sum, etc.)
df_yearly = df.groupby('Year')['pH (total scale)'].mean()
df_yearly = df.groupby('Year')['xCO2 Air (dry) (umol/mol)'].mean()


# Plotting the graph
sns.lineplot(data=df_yearly, marker='o', linestyle='-', color='b', label='Average pH (total scale)')
plt.title('Average pH (Total Scale) Over Years')
plt.xlabel('Year')
plt.ylabel('Average pH (Total Scale)')
plt.grid(True)
plt.show()

sns.lineplot(data=df_yearly, marker='o', linestyle='-', color='r', label='Average xCO2 Air (dry) (umol/mol)')
plt.title('Average xCO2 Air (dry) (umol/mol) Over Years')
plt.xlabel('Year')
plt.ylabel('Average xCO2 Air (dry) (umol/mol)')
plt.grid(True)
plt.show()
```
